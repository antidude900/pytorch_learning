

1.   For finding the whole dimension of a tensor(whether the tensor is 2D or 3D or ...?), see how many square brackets we have to go through to get to the items.

2.   For finding the size of a tensor on nth dimension(meaning if 2D dimension tensor, we have 1st dimenstion and 2nd dimension), see how many square it covers in that dimension.

3. For finding the index of a certain element or part of a tensor, keep in mind that the first element's/part's position is denoted as 0 meaning tensors are 0-indexed.

4. While indexing, we can use ":" meaning all the data inside that dimension are selected.

5. Unsqueezing means unpacking each element in seperate ranks(a rank for each element)[Rank means new row i.e each square bracket at the same dimension represents a rank]. It gives the feeling of unsqueezing as instead of squeezing all the elements in a single rank, we are spreading them in seperate ranks thus unsqueezing. Hence doing this, it adds a dimension in the shape of the tensor which is singleton meaning each rank in that dimension contains a single element.
